### 1. Prosty ranking wierzchołków
Zaimplementuj prosty model błądzenia przypadkowego po grafie skierowanym:

$$r(u)=c\sum_{v \in B_u} \frac{r(v)}{N_v} = c\sum_{v \in B_u} \frac{r(v)}{|F_v|}$$ 

gdzie $r(u)$ oznacza ranking wierzchołka $u$, parametr $c$ jest używany w normalizacji, $B_u$ jest zbiorem wierzchołków, z których wychodzą krawędzie do wierzchołka $u$, $F_v$ oznacza zbiór wierzchołków, do których dochodzą krawędzie z wierzchołka $v$, a $N_v = |F_v|$. W zapisie macierzowym:

$$r = cAr$$

gdzie $A$ jest macierzą adiacencji grafu, w której każdy wiersz $u$ jest przeskalowany wyjściowym stopniem wierzchołka $u$.

$$
A_{u,v} = \left\{ \begin{array}{ll}
\frac{1}{N_u} & \textrm{jeśli krawędź (u,v) istnieje}\\
0 & \textrm{w przeciwnym wypadku}
\end{array} \right.
$$

Zauważ, że $r$ może zostać obliczony jako dominujący wektor własny macierzy $A$ za pomocą metody potęgowej (dominujący wektor własny $q_1$ znormalizowany za pomocą normy $L1$). Przetestuj poprawność obliczeń korzystając z 3 dowolnych silnie spójnych grafów skierowanych o liczbie wierzchołków większej niż 10.

```python
def read_graph(file_name):
    G = nx.DiGraph(name='graph')
    file = open(file_name, "r")
    lines = file.read().split("\n")
        
    for line in lines:
        if line == '': continue
        points = line.split()
        G.add_edge(points[0], points[1])
    
    return G

def save_graph(G, file_name):
    nx.write_edgelist(G, file_name, data=False)

def draw_graph(G):
    if len(G) < 10:
        pos = nx.circular_layout(G)
        plt.figure(3,figsize=(5,5)) 
    else:
        pos = nx.spring_layout(G)
        plt.figure(3,figsize=(16,12)) 
        
    nx.draw_networkx_nodes(G, pos, node_color = 'purple', node_size = 400)
    nx.draw_networkx_labels(G, pos, font_size = 14, font_color = 'white')
    nx.draw_networkx_edges(G, pos, arrowsize=20)
    
def power(A, epsilon=1e-9, max_iteration=1000): #funkcja skopiowana z ostatniego sprawozdania (lab7)
    n = A.shape[0]
    x = np.random.uniform(size=(n, 1))
    
    for _ in range(max_iteration): # Obliczenia powinny zatrzymać sie po przekroczeniu maksymalnej liczby iteracji
        x_next = A @ x # Powtarzaj mnożenie wektora xi przez macierz A
        x_norm = x_next / np.linalg.norm(x_next, np.inf) # dzieląc za każdym razem wektor wynikowy przez norme
        
        if np.linalg.norm(x_norm - x) < epsilon: # Albo w przypadku gdy ||xi − xi+1|| < ε (kryterium małej poprawki)
            x = x_norm
            break
        
        x = x_norm # Przeskalowany wektor xi zbiega do dominującego wektora własnego
        
    return x / np.linalg.norm(x, 1) # Pod koniec obliczeń znormalizuj otrzymany wektor własny.

def convert_graph_to_matrix(G):
    A = nx.to_numpy_array(G)
    for index, u in enumerate(A): 
        if np.linalg.norm(u, 1) == 0: print(np.linalg.norm(u, 1))
        A[index] /= np.linalg.norm(u, 1)
        
    return A.T

def page_rank_v1(file_name):
    G = read_graph(file_name)
    A = convert_graph_to_matrix(G)
    draw_graph(G) 
    
    return power(A).flatten()
```

Testuję swój algorytm page_rank_v1 na 3 wierzchołkowym grafie z artykułu http://ilpubs.stanford.edu:8090/422/1/1999-66.pdf
<center><img src="graph1.png"></center>

```python
rank = page_rank_v1("graph1.txt")
library_rank = list(nx.pagerank(read_graph("graph1.txt"), alpha=1).values())

print("Wyliczone wartości:", rank)
print("Biblioteczne wartości:", library_rank)
print("Wartości zgadzają się:", np.allclose(rank, library_rank))
```

#### Przetestuj poprawność obliczeń korzystając z 3 dowolnych silnie spójnych grafów skierowanych o liczbie wierzchołków większej niż 10.

Na początku tworzę 3 silnie spójne grafy po to aby nie wystąpiła pętla tzw. "rank sink":
<center><img src="rank-sink.png"></center>

```python
def strongly_connected_graph(n, file_name):
    G = nx.fast_gnp_random_graph(n, 0.2, directed=True)
    while not nx.is_strongly_connected(G):
        G = nx.fast_gnp_random_graph(n, 0.2, directed=True)
        
    save_graph(G, file_name)
    
strongly_connected_graph(10, "graph2.txt")
strongly_connected_graph(20, "graph3.txt")
strongly_connected_graph(30, "graph4.txt")
```

### 2. Page Rank
Rozszerz model z poprzedniego zadania, dodając możliwość skoku do losowego wierzchołka grafu:
    
$$r(u)=c\sum_{v \in B_u} \frac{r(v)}{N_v} + (1-c)E(u) = c\sum_{v \in B_u} \frac{r(v)}{|F_v|} + (1-c)E(u) $$ 

W zapisie macierzowym:

$$r = (cA + (1-c)E \times 1)r$$

gdzie ${||r||}_{1} = 1$, a $E$ jest wektorem zawierającym prawdopodobieństwa odwiedzania wierzchołków przez losowy skok. Wykorzystaj metodę potęgową do obliczenia Page Rank jako dominującego wektora własnego macierzy $B = cA + (1-c)E \times 1$.

 1. $r_0$
 2. $do $
     3. $r_{i+1} = Br_i$ 
     4. $c = ||r_i||_1 − ||r_{i+1}||_1$ 
     5. $r_{i+1} = r_{i+1} + cE$
     6. $δ = ||r_{i+1} − r_i||_1$ 
 7.  $while\ δ > ε$
 
Przetestuj działanie zaimplementowanego algorytmu $Page Rank$ dla wybranych grafów z bazy SNAP. Przetestuj różne wartości parametru $c\ (1.0, 0.9, 0.8, 0.7, 0.6, 0.5)$ oraz różne postacie wektora $E$, przykładowo $E = \frac{1}{n}[1, 1, ... , 1]$.

```python
def normalize(v):
    norm = np.linalg.norm(v, ord = 1)
    if norm == 0: norm = np.finfo(v.dtype).eps
    return v / norm

def power_pagerank(A, c, E, epsilon, max_iteration=1000): 
    n = A.shape[0]
    x = normalize(np.random.uniform(size=(n, 1)))
    B = c * A + (1 - c) * (E * np.identity(n))
    
    for i in range(max_iteration): 
        x_next = B @ x 
        c = np.linalg.norm(x, 1) - np.linalg.norm(x_next, 1)
        x_next += c * E
        
        if np.linalg.norm(x_next - x, 1) < epsilon: 
            x = x_next
            break
        
        x = x_next 
        
    return x.flatten(), i

def page_rank_v2(file_name, c, E, draw=True, epsilon=1e-9):
    G = read_graph(file_name)
    A = normalize(nx.to_numpy_array(G))
    if draw: draw_graph(G) 
    
    return power_pagerank(A, c, E, epsilon)
```

Testuję swój algorytm page_rank_v2 na 3 wierzchołkowym grafie z artykułu.
$$E=[[0.(3), 0.(3), 0.(3)],
     [rand, rand, rand]],
     [0.5, 0.4, 0.1]] \ (znormalizowane)$$
$$C = [1.0, 0.9, 0.8, 0.7, 0.6, 0.5]$$

```python
E = [normalize(np.ones((3, 1))), normalize(np.random.uniform(size=(3, 1))), np.array([[0.5], [0.4], [0.1]])] 
C = [1.0, 0.9, 0.8, 0.7, 0.6, 0.5]

for (c, e) in [(x, y) for x in C for y in E]:
    rank, iterations = page_rank_v2("graph1.txt", c, e)
    library_rank = list(nx.pagerank(read_graph("graph1.txt"), alpha=1).values())
    print("c:", c, "iteracje:", iterations, "wyliczone:", rank, 
          "lib:", list(map(lambda x: round(x, 7), library_rank)))
```

Zauważam że:
 - Algorytm działa i zwraca poprawne wyniki zbliżone do bibliotecznych.
 - Czym większe $c$ tym dokładniejszy wynik, lecz również więcej iteracji potrzeba aby ten wynik obliczyć.
 - Najrozsądniejszym wyborem wydaje się być $E = \frac{1}{n}[1, 1, ... , 1]$
 
#### Przetestuj działanie zaimplementowanego algorytmu Page Rank dla wybranych grafów z bazy SNAP
Pobrałem grafy ze strony https://snap.stanford.edu/data/ego-Facebook.html oraz https://snap.stanford.edu/data/p2p-Gnutella08.html (BAZA SNAP). Przetestowałem działanie algorytmu na obu grafach.

```python
E = [normalize(np.ones((4039, 1))), normalize(np.random.uniform(size=(4039, 1)))]
C = [1.0, 0.9, 0.8, 0.7, 0.6, 0.5]

for (c, e) in [(x, y) for x in C for y in E]:
    rank, iterations = page_rank_v2("facebook.txt", c, e, False, 1e-10)
    print("E:", np.array(e[:3]).flatten())
    print("c:", c,"Maksymalna wartosc:", max(rank), "||r||:", sum(rank))
    print()
```

```python
E = [normalize(np.ones((6305, 1))), normalize(np.random.uniform(size=(6305, 1)))]
C = [1.0, 0.9, 0.8, 0.7, 0.6, 0.5]

for (c, e) in [(x, y) for x in C for y in E]:
    rank, iterations = page_rank_v2("p2p-Gnutella08.txt", c, e, False, 1e-10)
    print("E:", np.array(e[:3]).flatten())
    print("c:", c,"Maksymalna wartosc:", max(rank), "||r||:", sum(rank))
    print()
```

### Wnioski
 - Wnioskuje, że algorytm działa ze względu na $||r||_{1}=1$
 - Ulepszona wersja potrafi poradzić sobie z rank-sink'ami dzięki przeskokom które można kontrolować wartościami E oraz c. Gdy E dla danego wierzchołka jest duże to wynik końcowy zwiększy sie dla tego wierzchołka.
