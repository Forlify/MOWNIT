### 1.  Testowanie generatorów liczb losowych

#### Dla obydwu generatorów liczb losowych (Mersenne Twister oraz PCG64) oraz dla $n = 10, 1000, 5000$ wylosuj n liczb losowych pochodzących z rozkładu jednostajnego i wykonaj następujące kroki.

```python
ns = [10, 1000, 5000]

MT = []
PCG64 = []

for n in ns:
    MT.append(np.array([random.random() for _ in range(n)]))
    PCG64.append(np.random.default_rng().uniform(size=n))
```

1. Zwizualizuj na wykresie rozkład liczb w 10 równych przedziałach.

```python
def plot_histogram(data):
    fig, axs = plt.subplots(len(data), figsize=(10,10))
    
    for index, d in enumerate(data):
        axs[index].set_title('n = ' + str(d.shape[0]))
        axs[index].hist(d, color = 'r', edgeColor = 'black', bins=10)
        
    axs[-1].set_xlabel('Wartość')
    axs[1].set_ylabel('Ilość')
```

```python
print("Mersenne Twister:")
plot_histogram(MT)
```

```python
print("PCG64:")
plot_histogram(PCG64)
```

2. Sprawdź dla ilu liczb spełniona jest nierówność $x_i < x_{i+1}$. Ile powinno ich być dla idealnego generatora?

Dla idealnego generatora powinno byc $n/2$ spełnionych warunków.

```python
def count_balanced(data_array):
    for data in data_array:
        n = data.shape[0]
        count = sum([True if data[i] < data[i + 1] else False for i in range(n - 1)])
        print("Dla n =", n, "takich wartości jest:", count)
        
print("Mersenne Twister:")
count_balanced(MT)

print("\nPCG64:")
count_balanced(PCG64)
```

3. Zaimplementuj jeden z testów zdefiniowanych w rozdziale drugim artykułu https://csrc.nist.gov/publications/detail/sp/800-22/rev-1a/final i wykorzystaj go do sprawdzenia wylosowanego ciągu liczb. Czy widać różnice pomiędzy generatorami? Czy wraz z rosnącym n coś się zmienia?

Wybrałem algorytm Monobit. Moje wygenerowane liczby losowe są z zakresu $[0,1)$ a algorytm Monobit działa na liczbach binarnych $(0, 1)$. Z tego względu zastosowałem mapowanie zakresów: $[0, 1/2) \rightarrow 0, [1/2, 1) \rightarrow 1$    

```python
def monobit(data):
    count = 0
    
    for value in data:
        if value < 0.5: count -= 1
        else: count += 1

    s_obs = abs(count) / math.sqrt(len(data))
    return math.erfc(s_obs / math.sqrt(2))
```

```python
for i in range(len(MT)):
    print("Mersene Twister n =",len(MT[i]), "p_value =",monobit(MT[i]))
```

```python
for i in range(len(PCG64)):
    print("Permuted congruential generator n =",len(PCG64[i]), "p_value =", monobit(PCG64[i]))
```

Oba generatory mają wysoką jakość, z tego względu nie widać żadnych różnic pomiędzy generatorami. Test Monobit przeszły bez problemu. Wraz z rosnącym n nie zauważyłem zmian.

### 2. Generowanie liczb z rozkładu normalnego
#### Zaimplementuj generator liczb losowych pochodzących z rozkładu normalnego za pomocą metody Boksa-Mullera i zaprezentuj wyniki dla $n = 10, 100, 5000$ wygenerowanych liczb. Funkcja powinna dodatkowo przyjmować wartość oczekiwaną oraz odchylenie standardowe jako parametry.

```python
def box_muller(u1, u2, sigma = 1, mu = 0):
    z1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    z2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)
    return z1 * sigma + mu, z2 * sigma + mu
```

```python
xs = []
ys = []

for n in ns:
    u1 = np.random.random(n)
    u2 = np.random.random(n)
    
    x, y = box_muller(u1, u2)
    xs.append(x)
    ys.append(y)
    
    graph = sns.jointplot(x, y, kind='scatter', color="g")
    graph.fig.suptitle("n = " + str(n))
```

1. Zwizualizuj na wykresie rozkład liczb w 10 równych przedziałach. Na tym samym wykresie umieść funkcję gęstości prawdopodobieństwa rozkładu normalnego o takich samych parametrach.

```python
for x in xs:
    ax = sns.distplot(x, bins=10, kde=True, color='g')
    ax.set_title('n = ' + str(x.shape[0]))
    plt.show()
```

2. Wykonaj test statystyczny Shapiro-Wilka, przedstaw uzyskaną p-wartość, informację czy hipoteza zerowa została przyjęta/odrzucona dla $α = 0.05$ i czy w związku z tym są podstawy do stwierdzenia że liczby pochodzą z rozkładu normalnego.

```python
for x in xs:
    print("\nn =", x.shape[0])
    stat, p = shapiro(x)
    if p > 0.05:
        print('Próba pochodzi z populacji o rozkładzie normalnym, p =', p, 'stat =', stat)
    else:
        print('Próba nie pochodzi z populacji o rozkładzie normalnym, p =', p, 'stat =', stat)
```

Wyniki p > 5% nie oznaczają że hipoteza zerowa jest prawdziwa, jednak test oznacza, że jest to wysoce prawdopodobne. Przy wyżej otrzymanych wynikach można jednak stwierdzić, że liczby pochodzą z rozkładu normalnego.

Co ciekawe przyjmuje się, że test Shapiro-Wilka działa jedynie na małych próbach (nie większe niż pare tysięcy).

### 3. Metoda Monte Carlo
#### Napisz funkcję która zwróci przybliżoną wartość liczby pi wykorzystującą metodę Monte Carlo do obliczenia pola powierzchni koła jednostkowego. Funkcja powinna przyjmować liczbę wylosowanych punktów jako parametr. Zwizualizuj koło, zaznaczając na jednym kolorem punkty które trafiły do jego wnętrza, a drugim te które trafiły na zewnątrz. Na wykresie zwizualizuj błąd bezwzględny wraz z rosnącą liczbą punktów

```python
def absolute_error(real_value, estimated_value):
    return abs(real_value - estimated_value)

def point_in_circle(point):
    return point[0] ** 2 + point[1] ** 2 <= (1 / 2) ** 2

def random_point():
    return [0.5 - random.random(), 0.5 - random.random()]
    
def plot_estimation(points_in, points_out, number_of_points):
    fig, ax = plt.subplots(figsize=(16,16))
    ax.set_xlim(-1 / 2, 1 / 2)
    ax.set_ylim(-1 / 2, 1 / 2)
    
    points = list(zip(*points_in))
    ax.scatter(points[0], points[1], color='r', marker='.', label="Punkty w środku koła")
    
    points = list(zip(*points_out))
    ax.scatter(points[0], points[1], color='b', marker='.', label="Punkty poza kołem")
    
    ax.legend(loc="center")
    estimated_pi = 4 * len(points_in) / number_of_points
    ax.set(title = "%s punktów: %s w kole, przybliżone $\pi$ = %.4f. blad bezwzgledny = %.4f" % 
              (number_of_points, len(points_in), estimated_pi, absolute_error(np.pi, estimated_pi)))
    
    ax.add_artist(plt.Circle((0, 0), 0.5, fill=False, color='black'))
    
def pi_estimate(number_of_points):
    points_in = []
    points_out = []
    points_in_circle = 0
    
    for i in range(number_of_points):
        point = random_point();
        
        if point_in_circle(point):
            points_in.append(point)
            points_in_circle += 1
        else:
            points_out.append(point)
        
    plot_estimation(points_in, points_out, len(points_in) + len(points_out))
```

```python
pi_estimate(100)
```

```python
pi_estimate(1000)
```

```python
pi_estimate(100000)
```

```python
pi_estimate(1000000)
```

```python
pi_estimate(10000000)
```

2. Dokładność wyniku uzyskanego tą metodą jest zależna od liczby sprawdzeń i jakości użytego generatora liczb pseudolosowych. Zwiększanie liczby prób nie zawsze zwiększa dokładność wyniku, ponieważ generator liczb pseudolosowych ma skończenie wiele liczb losowych w cyklu.
